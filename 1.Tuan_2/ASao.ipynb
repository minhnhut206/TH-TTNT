Bài tập trên lớp

```python
class Graph:
    def __init__(self, adjac_lis):
        self.adjac_lis = adjac_lis

    def get_neighbors(self, v):
        return self.adjac_lis.get(v, [])

    def h(self, n):
        H = {
            'A': 1,
            'B': 1,
            'C': 1,
            'D': 1
        }
        return H[n]

    def a_star_algorithm(self, start, stop):
        open_lst = set([start])
        closed_lst = set([])

        g = {}
        g[start] = 0

        par = {}
        par[start] = start

        while len(open_lst) > 0:
            n = None

            for v in open_lst:
                if n is None or g[v] + self.h(v) < g[n] + self.h(n):
                    n = v

            if n is None:
                print("Không tồn tại đường đi!")
                return None

            if n == stop:
                reconst_path = []

                while par[n] != n:
                    reconst_path.append(n)
                    n = par[n]

                reconst_path.append(start)
                reconst_path.reverse()

                print("Tìm thấy đường đi:", reconst_path)
                return reconst_path

            for (m, weight) in self.get_neighbors(n):
                if m not in open_lst and m not in closed_lst:
                    open_lst.add(m)
                    par[m] = n
                    g[m] = g[n] + weight
                else:
                    if g[m] > g[n] + weight:
                        g[m] = g[n] + weight
                        par[m] = n

                        if m in closed_lst:
                            closed_lst.remove(m)
                            open_lst.add(m)

            open_lst.remove(n)
            closed_lst.add(n)

        print("Đường đi không tồn tại!")
        return None


# ============================
adjac_lis = {
    'A': [('B', 1), ('C', 3), ('D', 7)],
    'B': [('C', 1), ('D', 5)],
    'C': [('D', 12)]
}

graph1 = Graph(adjac_lis)
graph1.a_star_algorithm('A', 'C')
```

Bài tập về nhà

```python
class TSP:
    def __init__(self, distance_matrix, start=0):
        self.distance_matrix = distance_matrix
        self.start = start
        self.n = len(distance_matrix)

    def get_neighbors(self, state):
        current_city, visited_mask = state
        neighbors = []

        for next_city in range(self.n):
            if not (visited_mask & (1 << next_city)):
                new_mask = visited_mask | (1 << next_city)
                cost = self.distance_matrix[current_city][next_city]
                neighbors.append(((next_city, new_mask), cost))

        return neighbors

    def h(self, state):
        current_city, visited_mask = state

        unvisited = []
        for city in range(self.n):
            if not (visited_mask & (1 << city)):
                unvisited.append(city)

        if not unvisited:
            return self.distance_matrix[current_city][self.start]

        min_edges = 0
        for city in unvisited:
            min_dist = min(self.distance_matrix[city][j] for j in range(self.n) if j != city)
            min_edges += min_dist

        min_from_current = min(self.distance_matrix[current_city][city] for city in unvisited)

        return min_edges + min_from_current

    def a_star_tsp(self):
        start_state = (self.start, 1 << self.start)
        goal_mask = (1 << self.n) - 1

        open_lst = set([start_state])
        closed_lst = set([])

        g = {start_state: 0}
        par = {start_state: None}

        while len(open_lst) > 0:
            n = None

            for state in open_lst:
                if n is None or g[state] + self.h(state) < g[n] + self.h(n):
                    n = state

            if n is None:
                print("Không tồn tại đường đi!")
                return None

            current_city, visited_mask = n

            if visited_mask == goal_mask:
                total_cost = g[n] + self.distance_matrix[current_city][self.start]

                path = []
                while par[n] is not None:
                    path.append(n[0])
                    n = par[n]
                path.append(self.start)
                path.reverse()

                print(f"Tìm thấy đường đi: {path}")
                print(f"Tổng chi phí: {total_cost}")
                return path

            for (m, weight) in self.get_neighbors(n):
                if m not in open_lst and m not in closed_lst:
                    open_lst.add(m)
                    par[m] = n
                    g[m] = g[n] + weight
                else:
                    if g[m] > g[n] + weight:
                        g[m] = g[n] + weight
                        par[m] = n

                        if m in closed_lst:
                            closed_lst.remove(m)
                            open_lst.add(m)

            open_lst.remove(n)
            closed_lst.add(n)

        print("Đường đi không tồn tại!")
        return None


#================================
distance_matrix = [
    [0, 10, 15, 20],
    [10, 0, 35, 25],
    [15, 35, 0, 30],
    [20, 25, 30, 0]
]

tsp = TSP(distance_matrix, start=0)
tsp.a_star_tsp()
```
